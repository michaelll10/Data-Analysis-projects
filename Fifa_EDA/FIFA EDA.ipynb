```python
df['Age'].mean() #avarage age
```

```python
highest_one=df.Height.max() #max height wih name
df[['Name','Height']][df['Height'] == highest_one]
```

```python
lowest_weight=df.Weight.min() # min weight with name and club
df[['Name','Weight','Club']][df['Weight'] == lowest_weight]
```

```python
df[['Position']][df['Name']=='Cristiano Ronaldo'] #position of cristiano ronaldo
```

```python
#avarage value for each joined year
df.groupby('Joined')[['Value']].mean().sort_values(by='Value', ascending=False)
```

```python
#number of players for each club
df.groupby('Club')[['Name']].count().sort_values(by='Name', ascending=False ).rename(columns={'Name':'Count_Of_Players'}) 
```
